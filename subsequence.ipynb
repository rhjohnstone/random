```python
import numpy as np
from time import time
import matplotlib.pyplot as plt
%matplotlib inline

seed = 1
np.random.seed(seed)

def shortest_subsequence_adding_to(target, sequence):
    n = len(sequence)
    best_length = n
    for i in xrange(n):
        for j in xrange(i+1, i+1+best_length):
            if (np.sum(sequence[i:j])==target) and (j-i < best_length):
                best_length = j-i
                best_subseq = sequence[i:j]
                break
    try:
        return best_subseq
    except:
        print "No such subsequence found"
        return None
```

```python
limit = 100
target = 1
num_Ns = 40
num_times = 500
Ns = np.arange(1, num_Ns+1)*10
times = np.zeros(num_Ns)
for k, N in enumerate(Ns):
    start = time()
    offset = 0
    for _ in xrange(num_times):
        seqstart = time()
        sequence = np.random.randint(-limit, limit+1, N)
        seqtt = time()-seqstart
        offset += seqtt
        best_subsequence = shortest_subsequence_adding_to(target, sequence)
    time_taken = time()-start-offset
    times[k] = time_taken/num_times
plt.plot(Ns, times)
plt.plot()
```
